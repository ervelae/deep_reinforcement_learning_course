```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

fig, axs = plt.subplots(1,3, figsize=(15, 5))
colors = ["blue", "green", "red"]

for i, train_mass in enumerate([3,6,9]):
    scores = []
    masses = np.arange(3,9.5,0.5)
    for mass in masses:
        scores.append(np.load(f"total_rewards_mass{train_mass}_{mass}.npy"))
    scores = np.array(scores)
    scores_mean = scores.mean(axis=1)
    scores_perc90 = np.percentile(scores, 90, axis=1)
    scores_perc10 = np.percentile(scores, 10, axis=1)

    axs[i].plot(masses, scores_mean, label=f'm={train_mass}', color=colors[i])
    axs[i].fill_between(masses, scores_perc10.flatten(), scores_perc90.flatten(), alpha=0.3, label='10th-90th Percentile', color=colors[i])
    axs[i].set_title(f"Train mass: {train_mass}")
    axs[i].set_xlabel("Torso mass")
    if i==0:
        axs[i].set_ylabel("Performance")
    else:
        axs[i].set_yticklabels([])
    axs[i].legend()
    axs[i].set_ylim([0, 2000])
plt.show()
```
